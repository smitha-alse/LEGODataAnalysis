```python
import pandas as pd
df=pd.read_csv('C:\\Users\\smith\\Downloads\\parents_themes.csv')
df.dtypes
sets=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
#sets.shape[0]
```

```python
#sets[sets['set_num'].isnull()]
sets.dropna(subset = ['set_num'],inplace=True)
#sets.shape[0]
```

```python
df.isnull().any()
result_df= df.loc[df['is_licensed']==True]
result_df
```

```python
merged_df =  df.merge(sets,left_on = 'name',right_on = 'parent_theme')
#merged_df.dtypes
merged_df.drop(columns ='name_y',inplace=True)
merged_df.head()
```

```python
licensed = merged_df[merged_df['is_licensed']]
starwars = licensed[licensed['parent_theme']=='Star Wars']
Percentage_of_starwars = int(starwars.shape[0]/licensed.shape[0] *100)
Percentage_of_starwars
```

```python
licensed_sorted = licensed.sort_values('year')
#sum_groupbymax = licensed_sorted.groupby(['year','parent_theme']).size().groupby(['year']).max().reset_index(name='counts')
#sum_groupbymax.head(50)

sum_groupby = licensed_sorted.groupby(['year','parent_theme'])['is_licensed'].count().reset_index(name='counts')
sortedDF=sum_groupby.sort_values('counts', ascending=False).drop_duplicates('year')

new_era = sortedDF.max().year
new_era
```
